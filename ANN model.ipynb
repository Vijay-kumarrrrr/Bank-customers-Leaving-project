```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('BankCustomers.csv')
```

```python
df.head()
```

```python
data = pd.get_dummies(df, columns = ['Geography','Gender'])
data
```

```python
X = data.drop(['RowNumber','CustomerId','Surname','Exited'],axis = 1)
Y = data['Exited']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(activation="relu", input_dim=13, units=8, kernel_initializer="uniform"))
```

```python
# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=8, kernel_initializer="uniform"))
```

```python
# Adding the output layer
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy=accuracy_score(y_test,y_pred)
```

```python
cm
```

```python
accuracy*100
```

```python
from sklearn.metrics import classification_report 
clf_rep = classification_report(y_test, y_pred)
print(clf_rep)
```
